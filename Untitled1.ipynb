```python
import requests
from bs4 import BeautifulSoup
import re
import pandas as pd
url="https://www.flipkart.com/search?q=mobiles&as=on&as-show=on&otracker=AS_Query_TrendingAutoSuggest_1_0_na_na_na&otracker1=AS_Query_TrendingAutoSuggest_1_0_na_na_na&as-pos=1&as-type=TRENDING&suggestionId=mobiles&requestId=a6c983be-028c-46e0-9f18-9a33bb69dfa4"
```

```python
r =requests.get(url)
print(r.status_code)
```

```python
s=BeautifulSoup(r.text,"lxml")
```

```python
print(s)
```

```python
mobile = s.find("div",class_="_4rR01T")
```

```python
print(mobile)
```

```python
price=s.findAll("div",class_="_30jeq3 _1_WHN1")
print(len(price))
mobile_price=[]
for i in price:
    mobile_price.append(i.text)
print(mobile_price)
```

```python
mobiles = s.find_all("div",class_="_4rR01T")
print(len(mobiles))
mobile_names=[]
for i in mobiles:
    print(i.text)
    mobile_names.append(i.text)
```
